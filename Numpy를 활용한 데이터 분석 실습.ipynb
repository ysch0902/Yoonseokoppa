# [실습-1] BMI 지수 계산 및 과체중 이상인 BMI 값 확인
- 10명에 대한 키와 몸무게가 들어있는 'height_weight.txt'를 읽어 각 사랆별 BMI 지수를 구하고 비만 전 단계 이상인 사람들의 BMI값 확인

<pre>
    BMI 지수 = 몸무게(kg) / 키(M)의 제곱
    체질량지수의 범위 값에 따른 비만도 값
    
    18.5 미만 -> 저체중
    18.5 ~ 22.9 -> 정상
    23 ~ 24.9 -> 비만 전 단계
    25 ~ 29 -> 비만 1단계
    30 ~ 34.9 -> 비만 2단계
    35 이상 -> 고도비만
</pre>

```python
import numpy as np

data = np.loadtxt('height_weight.txt', delimiter=',') # 파일명, 구분자(어떤 문자로 구분할것인지)
print(data)
```

```python
# 키와 몸무게 데이터 추출

height = data[0]  * 0.01
print(height)
weight = data[1]
print(weight)
```

```python
# BMI 계산

bmi = np.divide(weight, height**2)
# bmi = weight / (height**2)
print(bmi)
```

```python
# 과체중 (BMI 23이상)인 사람들의 BMI 확인
# print(bmi[bmi >= 23])

과체중 = bmi[bmi >= 23]
print(과체중)

print(bmi[np.logical_and(bmi>=23, bmi<25)])
```

# [실습-2] 국가건강검진 혈압혈당 데이터 분석
- loadtxt() 함수 이용해 데이터 읽기: `skiprows = 1 파라미터` 추가해서 첫 행 읽기를 건너뜀
- 데이터 구조
    - 성별: 1 남자, 2 여자
    - 연령그룹: 1(20-24), 2(25-26), ...25~74세까지 2살 간격 1개 그룹으로 지정 27그룹(74세 이상)
    - 고혈압 당뇨병 진료여부: 1 고혈압 및 당뇨 진료 내역 있음, 2 고혈압 진료 내역 있음, 3 당뇨 진료 내역 있음, 4 고혈압 당뇨 진료내역 없음
    - 정상혈압: 수축기 혈압 120mmHg 미만, 확장기 혈압 80mmHg 미만
    - 공복혈당 수치: 126mg/dL 이상 -> 당뇨, 100mg/dL 미만 -> 정상
    
- 분석내용
    - 데이터 로딩, 전체 레코드 수와 상위 5개 데이터 확인, 데이터 모양 확인
    - 각 컬럼의 유니크 값 확인: (컬럼에 대한 unique 값 확인: unique() 함수 이용)
    - 여성과 남성의 데이터 수 확인
    - 여성과 남성의 수축기 혈압과 이완기 혈압의 평균값 확인
    - 정상 혈압 남자 및 여자 수 확인
    - 당뇨 질환(공복혈당 수치 126이상)을 가지고 있는 사람들의 평균 체질량 지수 확인

```python
# 데이터 로드 / 데이터 모양 및 수 확인
import numpy as np

data = np.loadtxt('국가건강검진_혈압혈당데이터.csv', delimiter=',', skiprows=1)

print(data.shape)
print('레코드 수: ', len(data))
print('상위 5개 데이터')
top5 = data[0:5, :] # 1행부터 5행까지 : 모든 열
print(top5)
```

```python
# 각 컬럼의 유니크 값 확인

print('성별 값: ', np.unique(data[:, 0]))
print('연령 값: ', np.unique(data[:, 1]))
print('고혈압 당뇨 진료여부 값: ', np.unique(data[:, 5]))
```

```python
# 여성과 남성의 데이터 수 확인
print('남성 데이터 수: ', len(data[data[:,0] == 1]))
print('여성 데이터 수: ', len(data[data[:,0] == 2]))
```

```python
# 여성과 남성의 수축기 혈압과 이완기 혈압의 평균값 확인
male_sbp_mean = np.mean(data[data[:,0]==1][:,2])
print("남성의 수축기 평균 혈압:%.2f"%male_sbp_mean)

male_dbp_mean = np.mean(data[data[:,0]==1][:,3])
print("남성의 이완기 평균 혈압:%.2f"%male_dbp_mean)

female_sbp_mean = np.mean(data[data[:,0]==2][:,2])
print("여성의 수축기 평균 혈압:%.2f"%female_sbp_mean)

female_dbp_mean = np.mean(data[data[:,0]==2][:,3])
print("여성의 이완기 평균 혈압:%.2f"%female_dbp_mean)
```

```python
# 정상혈압 남자 및 여자 수 확인
nbp_masking = np.logical_and(data[:,2]<120, data[:,3] < 80)

nbp = data[nbp_masking]
print("정상혈압 남자 수:", len(nbp[nbp[:,0]==1]))
print("정상혈압 여자 수:", len(nbp[nbp[:,0]==2]))
```

```python
# 당뇨 질환을 가지고 있는 사람들의 평균 BMI 값

disease = data[data[:,4]>=126]
print(disease[0:5,:])
#print(np.mean(disease[:,6]))
print(np.mean(disease[:, -1]))
```

# [실습-3] 시애틀 강수량 데이터 분석
- 데이터 로딩
- 강수령(PRCP)이 많은 날 상위 10일 확인
- 1월 강수량만 출력
- 1월에 내린 강수량의 합, 평균 확인
- 1년동안 강수량이 40미만으로 내린 날이 몇일인지?

```python
# 데이터 로딩
import numpy as np

data = np.loadtxt('Seattle2014.csv', delimiter=',', skiprows=1, dtype=object)
print(data[:5,:])
```

```python
# 강수량이 많은 날 상위 10일 확인

PRCP2 = data[:,2:4]
PRCP1 = np.array(PRCP2, dtype=int)
PRCP = np.sort(PRCP1, axis=0)[::-1]
print(PRCP[:10,:])
```

```python
# 쌤 버전
# 강수량이 많은 날 상위 10일 확인

prcp_data = data[:,3].astype(np.int32)
a = np.sort(prcp_data)[::-1]
print(a[:10])
```

```python
# 1월 강수량만 출력

jan = data[0:31, 2:4]
jan_np = np.array(jan, dtype=int)
print(jan_np)
```

```python
# 쌤 버전
# 1월 강수량만 확인

prcp_data = data[:,[2,3]]

month_prcp_data = np.array([ [date[4:6], prcp] for date, prcp in prcp_data], dtype= np.int32)
# print(month_prcp_data)

jan_prcp = month_prcp_data[month_prcp_data[:,0] == 1]
jan_prcp = jan_prcp[:,1]
print(len(jan_prcp))
print(jan_prcp)
```

```python
# 1월에 내린 강수량의 합, 평균

print(np.sum(jan_np[:,1], axis=0))
#print(np.average(jan_np, axis=0))
print(np.mean(jan_np[:,1], axis=0))
```

```python
# 쌤 버전
# 1월에 내린 강수량의 합, 평균

total_jan_prcp = np.sum(jan_prcp)
print("1월에 내린 강수량 합: %d"%total_jan_prcp)
avg_jan_prcp = np.mean(jan_prcp)
print("1월에 내린 강수량 평균: %.2f"%avg_jan_prcp)
```

```python
# 1년동안 강수량이 40미만으로 내린 날

data_np = np.array(data[:,2:4], dtype=int)
# print(data_np)
print(len(data_np[data_np[:,-1] < 40]))
print(data_np[data_np[:,-1] < 40])
```

```python
# 쌤 버전
# 1년동안 강수량이 40미만으로 내린 날

prcp_data = data[:,3:4].astype(int)
under_40 = prcp_data[prcp_data[:] < 40]
print(len(under_40))
print(under_40)
```

# [실습-4] 영화 평점 분석
- 데이터 로딩 (데이터 형식: user_id::item_id::rating::timestamp)
- 상위 10개 데이터 출력하기
- 데이터 정보 확인 (크기, 차원..)
- 전체 데이터 평균 평점 구하기
- 사용자 아이디 1인 데이터 출력
- 사용자 아이기 1인 사람의 평균 평점
- 각 사용자별 평균 평점(unique)
- 각 사용자별 평균 평점이 4점 이상인 사용자 및 인원 수 구하기

```python
# 데이터 로딩 및 상위 10개 데이터 출력, 데이터 정보 확인
import numpy as np

movie = np.loadtxt('ratings.dat', delimiter='::', dtype=int)
print(movie[:10,:])

print(movie.ndim)
print(movie.shape)
```

```python
# 전체 데이터 평균 평점 구하기

avg_rating = np.mean(movie[:,[2]])
print("평균평점: %.2f"%avg_rating)
```

```python
# 쌤 버전
# 전체 데이터 평균 평점 구하기

avg_ratings = np.mean(movie[:,2])
print("전체 사용자 평균평점: %.2f"%avg_ratings)
```

```python
# 사용자 아이디 1인 사람의 데이터 출력

user_1 = movie[movie[:, 0] == 1]

print(len(user_1))
print(user_1)
```

```python
# 사용자 아이디가 1인 사람의 평균 평점

user_1_rating = np.mean(user_1[:,2])
print("아이디가 1인 사람의 평균 평점: %.2f"%user_1_rating)
```

```python
# 각 사용자별 평균 평점 구하기

unique_user = np.unique(movie[:,:])
print(unique_user)
print(len(unique_user))

avg_uni_user = np.mean(unique_user[:,[2]])
print(avg_uni_user)
```

```python
# 각 사용자별 평균 평점

user_ids = np.unique(movie[:,0])
print(len(user_ids))

user_avg_rating = [[id, np.mean(movie[movie[:,0] == id][:,2])] for id in user_ids]
user_avg_rating = np.array(user_avg_rating, dtype=int)
print(user_avg_rating)
```

```python
print("평균 총점이 4점 이상인 유저의 수")
print(len(user_avg_rating[user_avg_rating[:,1] >= 4]), '명')
```
