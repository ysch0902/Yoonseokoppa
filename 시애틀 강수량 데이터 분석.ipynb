- 데이터 로딩
- 데이터 모양 및 정보 확인
- 필요하지 않은 컬럼 삭제 (STATION, STATION_NAME)
- DATE를 날짜 형식으로 수정
- 날짜 컬럼을 인덱스로 변경
- 강수량이 많은 날 상위 10일 출력
- 1월 강수량
- 1년 동안 강수령이 40 미만으로 내린 날의 수
- 1월에 내린 강수량의 합과 평균 강수량
- 2월에 내린 강수량의 합과 평균 강수량

```python
# 데이터 로딩

import numpy as np
import pandas as pd

data = pd.read_csv('seattle2014.csv')
display(data)
```

```python
# 데이터 모양 및 정보확인

print(data.shape)
print(data.info())
print(data.head(10))
```

```python
# 필요하지 않은 컬럼 삭제

data = data.drop(['STATION', 'STATION_NAME'], axis = 1)

data
```

```python
# DATE를 날짜 형식으로 수정

data.DATE = pd.to_datetime(data['DATE'].astype('str'))
data.head(5)
```

```python
# 날짜 컬럼을 인덱스로 변경

data
```
